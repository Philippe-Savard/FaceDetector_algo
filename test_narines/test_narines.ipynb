```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
image_name = 'narines6.jpg'
image = cv2.imread(image_name,cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (10,10))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

```python
for i,l in enumerate(image):
    for j,c in enumerate(l):
        if i < 200 or i > 300:
            image[i][j] = 255
        if j < 120 or j > 250:
            image[i][j] = 255
```

```python
plt.figure(figsize = (10,10))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

```python
for i,l in enumerate(image):
    for j,c in enumerate(l):
        if c > 40:
            image[i][j] = 255
```

```python
plt.figure(figsize = (10,10))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

```python
Gx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
Gy = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)
ForceGradient = np.sqrt(np.power(Gx,2)+np.power(Gy,2))
aretes = ForceGradient>90
plt.figure(figsize = (10,10))
plt.imshow(aretes, plt.get_cmap('binary'))
plt.show()
```

```python
# aretes = cv2.UMat(aretes)
new = np.float32(aretes)
new *= 255
```

```python
cv2.imwrite('aretes.png',new)
```
